```python
#Data cleaning

#rename
data=data.rename(columns={'Sr No':'S_number'})
#replace
data=data.replace("Not disclosed","TCS")
data=data.replace("#Fame","Rooman")

data=data.replace("Undisclosed","TATA")

data
```

```python
#Removes extra spaces from text data
data['investors']=data['investors'].str.strip()
data['startup']=data['startup'].str.strip()

data['vertical']=data['vertical'].str.strip()
data['city']=data['city'].str.strip()
data['state']=data['state'].str.strip()
data['round']=data['round'].str.strip()

data['investors']=data['investors'].str.strip()



data

```

```python
data['amount']=data['amount'].astype(int)

data
```

```python
#missing data
data.isnull()
data.isnull().sum()
data.isnull().sum().sum()

data['investors']=data['investors'].fillna(data['investors']).mode()[0]
data


```

```python
import matplotlib.pyplot as plt
#1.how has strtup funding trend change over the years?
funding_trend=data.groupby('year')['amount'].sum().reset_index()
plt.figure(figsize=(10,6))
plt.plot(funding_trend['year'],funding_trend['amount'],marker='o',linestyle='--',color="blue")
plt.title('Startup Funding Trend Over the Years',color="black")
plt.xlabel('Year')
plt.ylabel('Total Funding Amount')
plt.grid(True)
plt.show()
```

```python
from matplotlib.lines import lineStyles
#2.which cities have the highest number of startup
top_cities=data['city'].value_counts().head(10)
color =["blue","black","green","Yellow","purple","orange","brown","skyblue","pink","red"]
plt.figure(figsize=(10,6))
plt.bar(top_cities.index,top_cities.values,color=color,alpha=0.4,edgecolor="black")
plt.xlabel('City')
plt.ylabel('Number of Startups')
plt.title('Top 10 Cities with the Highest Number of Startups')
plt.xticks(rotation=45)
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#3.Which startup sectors have the heihest number of funded startups?
#bar chart
#horigental
import matplotlib.pyplot as plt
Demo=data["vertical"].value_counts().head(10)
color=["red","blue","skyblue","yellow","brown","pink","black","orange","green","purple"]
plt.figure(figsize=(10,6))
plt.barh(Demo.index,Demo.values,color=color,alpha=0.4,edgecolor="black")
plt.xlabel('Number of Startups')
plt.ylabel('Startup Sector')
plt.title('Top 10 Startup Sectors with the Highest Number of Funded Startups')
plt.grid(axis='x',linestyle='--',color="grey")
plt.tight_layout()
plt.show()
```

```python
#4.what is the distribution of different funding rounds.?
import matplotlib.pyplot as plt
funding_round=data['round'].value_counts().head(7)
color=["red","blue","skyblue","yellow","brown","pink","black","orange","green","purple"]
plt.figure(figsize=(10,6))
plt.pie(funding_round,labels=funding_round.index,autopct='%1.1f%%',colors=color)
plt.title('Round Distribution of Different Funding Rounds')
```

```python
#how are starting
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.hist(data['amount'], bins=30, color="skyblue", edgecolor="black", alpha=0.7)
plt.xlabel('Funding Amount')
plt.ylabel('Frequency')
plt.title('Distribution of Funding Amounts')
plt.grid(axis='y',linestyle='--',color="grey")
plt.tight_layout()
plt.show()
```

```python
#6. The relationship between funding year and funding amount?
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(data['year'],data['amount'],color="blue",alpha=0.7)
plt.xlabel('Funding Year')
plt.ylabel('Funding Amount')
plt.title('The funding amount by year')
plt.grid(axis='y',linestyle='--',color="grey")
plt.show()

```

```python
#7.Which funding round occurs most frequently?
import matplotlib.pyplot as plt
import seaborn as sns
# Get the top 10 funding rounds by frequency
round_counts = data['round'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.countplot(data=data, x='round', order=round_counts.index, palette="viridis")
plt.xlabel('Funding Round')
plt.ylabel('Frequency')
plt.title('Funding Round Distribution')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
